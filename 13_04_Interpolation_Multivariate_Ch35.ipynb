# Mutivariate Interpolation 

### with scipy.interpolate import griddata

```python
from scipy.interpolate import griddata
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x,y):
    return np.cos(x) + np.sin(y)

# define the full grid
xg, yg = np.mgrid[np.pi:3*np.pi:100j, np.pi:3*np.pi:100j]

# get points on the grid
xpts = np.linspace(np.pi, 3*np.pi,10)
ypts = np.linspace(np.pi, 3*np.pi,10)
pts  = np.array([[x,y] for y in ypts for x in xpts])
fpts = f(pts[:,0], pts[:,1])
```

```python
# do a cubic interpolation
fit = griddata(pts, fpts, (xg, yg), method = "cubic")

# do a nearest interpolation
fitn = griddata(pts, fpts, (xg, yg), method = "nearest")
```

```python
fig,(ax1, ax2, ax3) = plt.subplots(1,3)

#plot original function
p1 = ax1.imshow(f(xg,yg).T, extent = [np.pi, 3*np.pi, np.pi, 3*np.pi], origin="lower")
ax1.set_title("Original Function")

c1 = ax2.imshow(fit.T, extent = [np.pi, 3*np.pi, np.pi, 3*np.pi], origin="lower")
ax2.set_title("Cubic Spline Function")

v1 = ax3.imshow(fitn.T, extent = [np.pi, 3*np.pi, np.pi, 3*np.pi], origin="lower")
ax3.set_title("Narest Function")
```
